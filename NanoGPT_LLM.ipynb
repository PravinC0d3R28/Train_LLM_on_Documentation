View and prepare the data

```python
import pandas as pd

df = pd.read_csv("/content/doc.csv")

df = df[::-1]

with open("extracted_data.txt", "w") as f:
    for index, row in df.iterrows():
        insel_data = str(row['insel'])
        f.write(insel_data + "\n")

        textpart_data = str(row['textpart'])
        f.write(textpart_data + "\n")

        f.write("\n")

import re
with open("extracted_data.txt", "r") as f:
    extracted_data = f.read()

pattern = re.compile(r"Found a content problem with this page\?Edit the page on GitHub\..*?This page was last modified on .*? by MDN contributors\.", re.DOTALL)
cleaned_data = re.sub(pattern, "", extracted_data)

with open("cleaned_data.txt", "w") as f:
    f.write(cleaned_data)
```
